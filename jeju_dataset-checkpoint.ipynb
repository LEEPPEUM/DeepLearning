```python
import pandas as pd
bc_card = pd.read_csv('./dummies/201901-202003.csv')
bc_card = bc_card.dropna()

jeju = bc_card[bc_card['CARD_SIDO_NM'] == '제주']

y = jeju['AMT']
x = jeju[['REG_YYMM', 'STD_CLSS_NM', 'CARD_SIDO_NM', 'HOM_SIDO_NM', 'AGE', 'SEX_CTGO_CD', 'FLC']]

x_dummy = pd.get_dummies(x, columns = ['REG_YYMM', 'STD_CLSS_NM', 'CARD_SIDO_NM', 'HOM_SIDO_NM', 'AGE', 'SEX_CTGO_CD', 'FLC'], 
                                       drop_first = True)  # one-hot

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_dummy, y, test_size=0.2)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2)
```
